```python
import os
from langchain_community.document_loaders import (
    UnstructuredPDFLoader
)

docs_folder = "./Docs"
documents = []

for filename in os.listdir(docs_folder):
    filepath = os.path.join(docs_folder, filename)
    loader = UnstructuredPDFLoader(filepath)
    documents.extend(loader.load())
```

Chunking the text into smaller parts can be useful for processing large documents. Here's how you can do it using Python:

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = splitter.split_documents(documents)

for i, chunk in enumerate(chunks):
    print(f"------ Chunck {i + 1} ------ ")
    print(chunk.page_content)
    print("------------------------------")

# - chunk_size=300: Each chunk will be up to 300 characters long.
# - chunk_overlap=50: Each chunk will overlap the previous one by 50 characters to maintain context.
# - Splitting documents into smaller chunks is important because it allows language models to process the information more efficiently, 
# improving speed and effectiveness while also addressing the token limits often imposed by these models. 
# By working with smaller segments, models can generate responses more quickly and accurately.
```
