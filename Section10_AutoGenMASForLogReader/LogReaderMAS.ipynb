```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager
import os
llm_config = {
    "config_list": [{
        "model": "qwen2.5:latest",
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama"
    }]
}


llm_config_for_manager = {
    "config_list": [{
        "model": "qwen3:8b",
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama"
    }]
}
```

#### Read the Log files

```python
def read_logs():
    folder = "./logs"
    logs = []
    
    if not os.path.exists(folder):
        return "Error the .log path is missing"
    
    for file in os.listdir(folder):
        with open(os.path.join(folder, file), "r") as f:
            logs += f"\n--- {file} ---\n{f.read()}"
            
    return logs if logs else "No log content found"
```

```python
log_anlayst = AssistantAgent(
    
    name = "LogAnalyser",
    system_message="""
    
    You are a QA Engineer who knows how to find errors in the logs.
    Analyse the logs for:
    - Errors
    - Warns
    - Permission errors
    - Critial issue
    
    Format:
    Log Summary
    - Errors: [#]
    - Warns: [#]
    - Permission errors: [#]
    - Critial issue: [#]
    
    End with: 'Analysis Complete.'
    """,
    llm_config=llm_config
)
```

#### User Proxy Agent

```python
qa_lead = UserProxyAgent(
    name="QALead",
    human_input_mode="NEVER",
    code_execution_config=False,
    system_message="You are the QA Lead running this log analysis process"
)
```

```python
logs = read_logs()

# Team collaboration
chat = GroupChat(agents=[qa_lead, log_anlayst], max_round=6, speaker_selection_method='round_robin')
manager = GroupChatManager(groupchat=chat, llm_config=llm_config_for_manager)

#start the conversion
message =   f"""
            Please analyse the following logs: {logs}
            Identify issues and suggest fixes if there is any.
            """
            
qa_lead.initiate_chat(manager, message=message)
```

- speaker_selection_method='round_robin'
    - Defines how the speaking agent is chosen each turn.
    - round_robin → agents speak in order:
        - qa_lead → log_anlayst → qa_lead → log_anlayst → ...
    - Other possible strategies (depending on version):
        -"random" → random agent speaks
        - "manual" → you choose dynamically
        - "auto" → LLM decides the next speaker
