### Building Test Case Generator with MultiAgent Systems using AutoGen

````python
# !pip install autogen
# !pip install "autogen-ext[ollama]"
````

````python
from autogen import AssistantAgent, UserProxyAgent


llm_config = {
    "config_list": [{
        "model": "qwen2.5:latest",
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama"
    }]
}
````

#### Create an Assistant Agent

```python

````python
qa_assistant = AssistantAgent(
    name="QAAssistant",
    description="This agent is is for QA operation",
    system_message="You are a helpful QA assistant who can write BDD scenarios. You create not more than test cases asked by user",
    llm_config=llm_config
)
````

#### Creating User Proxy Agent

```python

````python
qa_lead = UserProxyAgent(
    name="QALead",
    human_input_mode="NEVER",
    system_message="You are the QA Lead asking questions.",
    code_execution_config=False
)
````

#### Initiate the Chat !

````python
qa_lead.initiate_chat(
    recipient=qa_assistant,
    max_turns=3,
    message="Can you help me write atleast 10 BDD Test cases for Login and forget password"
)
````
