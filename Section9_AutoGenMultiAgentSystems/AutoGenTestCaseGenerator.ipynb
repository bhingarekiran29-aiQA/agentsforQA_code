### Building Test Case Generator with MultiAgent Systems using AutoGen

```python
# %pip install autogen
# %pip install "autogen-ext[ollama]"
# %pip install openai
```

```python
from autogen import AssistantAgent, UserProxyAgent

llm_config = {
    "config_list": [{
        "model": "qwen2.5:latest",
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama"
    }]
}
# You’re using /v1 only because Autogen requires an OpenAI-style API, and the proxy in front of Ollama provides that. You’re not using the native Ollama API here.
```

#### Create an Assistant Agent

```python
qa_assistant = AssistantAgent(
    name="QAAssistant",
    description="This agent is is for QA operation",
    system_message="You are a helpful QA assistant who can write BDD scenarios. You create not more than test cases asked by user",
    llm_config=llm_config
)
```

#### Creating User Proxy Agent

```python
qa_lead = UserProxyAgent(
    name="QALead",
    human_input_mode="NEVER",
    system_message="You are the QA Lead asking questions.",
    code_execution_config=False
)
```

##### human_input_mode="NEVER"
- This agent will not wait for human input.
- It will automatically send messages without stopping for user responses.

- Values:
    - "ALWAYS" → Always wait for human input
    - "NEVER" → Never ask, fully autonomous
    - "TERMINATE" → Stops after each message

##### code_execution_config=False
- Disables code execution for this agent.
- Means this agent cannot run Python code or execute scripts.
- It will only generate text.

#### Initiate the Chat !

```python
qa_lead.initiate_chat(
    recipient=qa_assistant,
    max_turns=3,
    message="Can you help me write atleast 10 BDD Test cases for Login and forget password"
)
```

- max_turns=3
    - Maximum number of message exchanges (turns) between the agents.
    - A “turn” = one message from initiator + one response from recipient.
    - Prevents infinite loops.
