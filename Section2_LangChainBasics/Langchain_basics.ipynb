### Setting up ollama for LangChain with Model qwen-8b

````python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    base_url="http://localhost:11434",
    model="llama3.2:latest",
    temperature=0.5
)
````

#### Call the ollama model

````python
response = llm.invoke("Hi, How are you doing ?")
print(response)  
print(response.content)

# So the LLM will give the response as : content = "response", hence to read the response only, used response.content.
````

### Streaming the response

````python
for chuck in llm.stream("Hi How are you doing ?"):
    print(chuck.content, end="", flush=True)
````

| Aspect             | `invoke()`                          | `stream()`                      |
| ------------------ | ----------------------------------- | ------------------------------- |
| Response type      | Full response at once               | Incremental (token/chunk-wise)  |
| Blocking           | Waits until completion              | Starts returning immediately    |
| Output             | Single `AIMessage`                  | Generator of chunks             |
| Latency visibility | No                                  | Yes (real-time)                 |
| Best for           | Validation, assertions, batch tests | UI, chat apps, live feedback    |
| Testing complexity | Simple                              | Higher (ordering, completeness) |

- invoke() is for correctness; stream() is for experience.
- flush=True ensures streamed LLM tokens are printed immediately by bypassing output buffering, which is essential for real-time response display

#### Understanding Prompt Templates

````python
from langchain_core.prompts import PromptTemplate
prompt_template = PromptTemplate.from_template("Hi how are you, I am {name}")
print(llm.invoke(prompt_template.invoke({'name': 'Kiran'})).content)
# This code demonstrates prompt parameterization using LangChain’s PromptTemplate, enabling reusable and 
# testable prompts that are dynamically populated and sent to an LLM.
````

````python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template("What is the role of {type} in software Testing")

prompt = prompt_template.invoke({'type': 'Human'})
print(prompt) # -> What is the role of Human in software Testing
# prompt = prompt_template.invoke({}) - we can pass it empty if no parametrization is needed. 

result = llm.invoke(prompt)

print(result.content)

# this prompt template is reusable as compared to above in line 6.
# markdownlivepreview.com can be used to view the result in good format.
````

#### LangChain Chaining mechanism

```python

````python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template("What is the role of {type} in software Testing")
chain = prompt_template | llm  # only combining above code in chain to make it short. 
result = chain.invoke({'type': 'AI'})

print(result.content)
````

#### Chaining with Parsers

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt_template = PromptTemplate.from_template("What is the role of {type} in software Testing")
chain = prompt_template | llm | StrOutputParser() # this will change the response of LLM to String
result = chain.invoke({'type': 'AI'})

print(result)
````

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser

prompt_template = PromptTemplate.from_template("What is the role of {type} in software Testing ? give me the response in JSON format ONLY")
chain = prompt_template | llm | JsonOutputParser() # it will give the response in JSON format, for that prompt need to be updated to ask for the response in JSON format. 
result = chain.invoke({'type': 'AI'})

print(result)
````
